```python
input_tab = Table.read('Data/Monthly_CP_Index_1980-2020.txt',format='ascii')
yearList=[]
janList=[]
febList=[]
marList=[]
aprList=[]
mayList=[]
junList=[]
julList=[]
augList=[]
sepList=[]
octList=[]
novList=[]
decList=[]
for i in range(1980,2021):
    yearList.append(i)
index_value=input_tab['col5']

count=0
for elem in input_tab['col1']:
    if elem[3:6] == 'JAN':
        janList.append(input_tab['col5'][count])
    elif elem[3:6] == 'FEB':
        febList.append(input_tab['col5'][count])
    elif elem[3:6] == 'MAR':
        marList.append(input_tab['col5'][count])
    elif elem[3:6] == 'APR':
        aprList.append(input_tab['col5'][count])
    elif elem[3:6] == 'MAY':
        mayList.append(input_tab['col5'][count])
    elif elem[3:6] == 'JUN':
        junList.append(input_tab['col5'][count])
    elif elem[3:6] == 'JUL':
        julList.append(input_tab['col5'][count])
    elif elem[3:6] == 'AUG':
        augList.append(input_tab['col5'][count])
    elif elem[3:6] == 'SEP':
        sepList.append(input_tab['col5'][count])
    elif elem[3:6] == 'OCT':
        octList.append(input_tab['col5'][count])
    elif elem[3:6] == 'NOV':
        novList.append(input_tab['col5'][count])
    elif elem[3:6] == 'DEC':
        decList.append(input_tab['col5'][count])
    else:
        continue
    #print(input_tab['col5'][count])
    count=count+1
    #print(elem)
#print(len(yearList),len(janList))
janList=janList[:-1]  #remove last item from janList because its the only month with 2021 data
output_tab=Table([yearList,janList,febList,marList,aprList,mayList,junList,julList,augList,sepList,octList,novList,decList],
                 names=('Year','Jan','Feb','Mar','Apr','May','June','July','Aug','Sep','Oct','Nov','Dec'))
print(output_tab)
output_tab.write('CP_index_1980-2020.csv',overwrite=True)
output_tab.show_in_notebook()
```

```python
tab=Table.read('CP_index_1980-2020.csv',format='ascii')
monthList=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
djfList=[]
jfmList=[]
fmaList=[]
mamList=[]
amjList=[]
mjjList=[]
jjaList=[]
jasList=[]
asoList=[]
sonList=[]
ondList=[]
ndjList=[]
yearList=[]
for i in range(len(tab['Year'])):
    if tab['Year'][i] == 1979:
        continue
    elif tab['Year'][i] == 2020:
        djf = tab['Dec'][i-1]+tab['Jan'][i]+tab['Feb'][i]
        jfm = tab['Jan'][i]+tab['Feb'][i]+tab['Mar'][i]
        fma = tab['Feb'][i]+tab['Mar'][i]+tab['Apr'][i]
        mam = tab['Mar'][i]+tab['Apr'][i]+tab['May'][i]
        amj = tab['Apr'][i]+tab['May'][i]+tab['June'][i]
        mjj = tab['May'][i]+tab['June'][i]+tab['July'][i]
        jja = tab['June'][i]+tab['July'][i]+tab['Aug'][i]
        jas = tab['July'][i]+tab['Aug'][i]+tab['Sep'][i]
        aso = tab['Aug'][i]+tab['Sep'][i]+tab['Oct'][i]
        son = tab['Sep'][i]+tab['Oct'][i]+tab['Nov'][i]
        ond = tab['Oct'][i]+tab['Nov'][i]+tab['Dec'][i]
        ndj = tab['Nov'][i]+tab['Dec'][i]  
        
        djfAvg = np.round(djf/3,2)
        jfmAvg = np.round(jfm/3,2)
        fmaAvg = np.round(fma/3,2)
        mamAvg = np.round(mam/3,2)
        amjAvg = np.round(amj/3,2)
        mjjAvg = np.round(mjj/3,2)
        jjaAvg = np.round(jja/3,2)
        jasAvg = np.round(jas/3,2)
        asoAvg = np.round(aso/3,2)
        sonAvg = np.round(son/3,2)
        ondAvg = np.round(ond/3,2)
        ndjAvg = np.round(ndj/2,2)
        
        yearList.append(tab['Year'][i])
        djfList.append(djfAvg)
        jfmList.append(jfmAvg)
        fmaList.append(fmaAvg)
        mamList.append(mamAvg)
        amjList.append(amjAvg)
        mjjList.append(mjjAvg)
        jjaList.append(jjaAvg)
        jasList.append(jasAvg)
        asoList.append(asoAvg)
        sonList.append(sonAvg)
        ondList.append(ondAvg)
        ndjList.append(ndjAvg)
    else:
        djf = tab['Dec'][i-1]+tab['Jan'][i]+tab['Feb'][i]
        jfm = tab['Jan'][i]+tab['Feb'][i]+tab['Mar'][i]
        fma = tab['Feb'][i]+tab['Mar'][i]+tab['Apr'][i]
        mam = tab['Mar'][i]+tab['Apr'][i]+tab['May'][i]
        amj = tab['Apr'][i]+tab['May'][i]+tab['June'][i]
        mjj = tab['May'][i]+tab['June'][i]+tab['July'][i]
        jja = tab['June'][i]+tab['July'][i]+tab['Aug'][i]
        jas = tab['July'][i]+tab['Aug'][i]+tab['Sep'][i]
        aso = tab['Aug'][i]+tab['Sep'][i]+tab['Oct'][i]
        son = tab['Sep'][i]+tab['Oct'][i]+tab['Nov'][i]
        ond = tab['Oct'][i]+tab['Nov'][i]+tab['Dec'][i]
        ndj = tab['Nov'][i]+tab['Dec'][i]+tab['Jan'][i+1]
            
        djfAvg = np.round(djf/3,2)
        jfmAvg = np.round(jfm/3,2)
        fmaAvg = np.round(fma/3,2)
        mamAvg = np.round(mam/3,2)
        amjAvg = np.round(amj/3,2)
        mjjAvg = np.round(mjj/3,2)
        jjaAvg = np.round(jja/3,2)
        jasAvg = np.round(jas/3,2)
        asoAvg = np.round(aso/3,2)
        sonAvg = np.round(son/3,2)
        ondAvg = np.round(ond/3,2)
        ndjAvg = np.round(ndj/3,2)
        
        yearList.append(tab['Year'][i])
        djfList.append(djfAvg)
        jfmList.append(jfmAvg)
        fmaList.append(fmaAvg)
        mamList.append(mamAvg)
        amjList.append(amjAvg)
        mjjList.append(mjjAvg)
        jjaList.append(jjaAvg)
        jasList.append(jasAvg)
        asoList.append(asoAvg)
        sonList.append(sonAvg)
        ondList.append(ondAvg)
        ndjList.append(ndjAvg)
anomaly_tab=Table([yearList,djfList,jfmList,fmaList,mamList,amjList,mjjList,jjaList,jasList,asoList,sonList,ondList,ndjList],
                  names=('Year','djf','jfm','fma','mam','amj','mjj','jja','jas','aso','son','ond','ndj'))      
#print(anomaly_tab)
anomaly_tab.show_in_notebook()
anomaly_tab.write('CP_Anomalies_1980-2020.csv',overwrite=True)
```
